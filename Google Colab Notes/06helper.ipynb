```python
file_path = '/content/drive/MyDrive/SustainabilityReports/Firm_ID/Results/DB/esg_report.csv'
```

```python

df = pd.read_csv(file_path)

def clean_company_name(name):
    name = name.replace('_ESG_EN', '').replace('_IR_EN', '')
    name = name.replace('_', ' ')
    return name

df['Company'] = df['Company'].apply(clean_company_name)

updated_file_path = '/content/drive/MyDrive/SustainabilityReports/Firm_ID/Results/DB/esg_report.csv'
df.to_csv(updated_file_path, index=False)

print(f"Updated file saved to: {updated_file_path}")
```

```python
indicator_file_path = '/content/drive/MyDrive/SustainabilityReports/Firm_ID/Results/DB/indicator_counts.csv'
```

```python
df = pd.read_csv(indicator_file_path)

columns_to_sum = df.columns[df.columns.get_loc('Published Year') + 1:]

df['qualitative_sentences'] = df[columns_to_sum].sum(axis=1, numeric_only=True)

output_file_path = '/content/drive/MyDrive/SustainabilityReports/Firm_ID/Results/DB/indicator_counts_with_sum.csv'
df.to_csv(output_file_path, index=False)

print(f"Updated file saved to {output_file_path}")
```
